```python
from bs4 import BeautifulSoup
soup = BeautifulSoup()
stopwords = []
for i in stop:
    x = i.replace("\n",'')
    stopwords.append(x)
```

```python
with open('ratings_train.txt',mode= 'r',encoding= 'UTF-8') as f:
    example = f.readlines()
```

```python
def spacing_okt(wrongSentence):
    tagged = okt.pos(wrongSentence)
    corrected = ""
    for i in tagged:
        if i[1] in ('Josa', 'PreEomi', 'Eomi', 'Suffix', 'Punctuation'):
            corrected += i[0]
        else:
            corrected += " "+i[0]
    if corrected[0] == " ":
        corrected = corrected[1:]
    return corrected

print(spacing_okt("대한민국은민주공화국이다.대한민국의모든권력은국민으로부터나온다."))
```

```python
from bs4 import BeautifulSoup
sentences = []
labels = []
for items in example:
    sentence = items.replace(',', ' , ')
    sentence = sentence.replace('.', ' . ')
    sentence = sentence.replace('-', ' - ') 
    sentence = sentence.replace('/', ' / ')
    label = sentence.split('\t')[0]
    labels.append(label) 
    sentences.append(sentence.split('\t')[1:-1])
```

```python
list_sentences = []
for i in sentences:
    for j in i:
        # x = spacing_okt(j)
        list_sentences.append(j)
```

```python
new_sentence = []
for i in list_sentences:
    if len(i) != 0:
        new_sentence.append(i)
```

```python
space_sentences = []
for i in new_sentence:
    x = spacing_okt(i)
    space_sentences.append(x)    
```

```python
import string
table = str.maketrans('', '',string.punctuation)
sentences = []
for i in space_sentences:
    sentence = i.replace(',', ' , ')
    sentence = sentence.replace('.', ' . ')
    sentence = sentence.replace('-', ' - ') 
    sentence = sentence.replace('/', ' / ')
    words = sentence.split()
    filtered_sentence = ""
    for word in words:
        word = word.translate(table)
        if word not in stopwords:
            filtered_sentence = filtered_sentence + word + " "
    sentences.append(filtered_sentence)
```

```python
last_sentence =[]
for i in sentences:
    try:
        x1= i.replace(' ','')
        x2 = spacing_okt(x1)
        last_sentence.append(x2)
    except:
        pass
```

```python
train_size = 120000
train_sentences = last_sentence[0:train_size]
test_sentences = last_sentence[train_size:]
```
