### KNN 기반 모델
#### [1] 분류 - 품종 분류

```python
#모듈 로딩
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
#데이터 로딩
data = pd.read_csv('../dataset/sample_data.csv')
data.head()
```

```python
#결측치 확인
data.info()
print()
print(data.isnull().sum())
print()
data.describe()
#데이터간 범위가 너무 큼
```

```python
#스케일링
mms = MinMaxScaler()
mms_data = mms.fit_transform(data[data.columns[:2]])
mms_data = pd.DataFrame(mms_data, columns= ['Length','Weight'])
mms_data
```

```python
#종 데이터 category화
data['Species'] = data['Species'].astype('category')
data.info()
```

```python
featue_data = mms_data
target_data = data['Species']
X_train, X_test, y_train, y_test = train_test_split(featue_data, target_data, test_size=0.2, random_state=42)

param = {'n_neighbors': [i for i in range(1, 50)], 'weights': ['uniform', 'distance'],'p':[1,2,3]}
knn_model = KNeighborsClassifier()

Knn_grid = GridSearchCV(knn_model, param_grid=param, cv=5,refit=True)
Knn_grid.fit(X_train, y_train)

```

```python
print(f'Best parameters: {Knn_grid.best_params_}')
print(f'Best score: {Knn_grid.best_score_}')
best_model_of_knnclf = Knn_grid.best_estimator_
pred = best_model_of_knnclf.predict(X_test)
print(f'Accuracy: {accuracy_score(pred,y_test)}')
pd.DataFrame(Knn_grid.cv_results_)
```

#### [2] 회귀 길이로 무게 측정

```python
#모듈 로딩
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
featue_data = np.array(mms_data['Length']).reshape(-1,1)
target_data = np.array(data['Weight']).reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(featue_data, target_data, test_size=0.2, random_state=42)

param = {'n_neighbors': [i for i in range(1, 50)], 'weights': ['uniform', 'distance'],'p':[1,2,3]}
knn_model = KNeighborsRegressor()

Knn_grid = GridSearchCV(knn_model, param_grid=param, cv=5,refit=True)
Knn_grid.fit(X_train, y_train)

```

```python
print(f'Best parameters: {Knn_grid.best_params_}')
print(f'Best score: {round(Knn_grid.best_score_,3)}')
best_model_of_knnclf = Knn_grid.best_estimator_
pred = best_model_of_knnclf.predict(X_test)
print(f'best model: {best_model_of_knnclf}')
pd.DataFrame(Knn_grid.cv_results_)
```
